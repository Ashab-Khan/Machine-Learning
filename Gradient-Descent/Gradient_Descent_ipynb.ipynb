```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,Y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=5)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
```

```python
#Lets apply OLS
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,Y)
plt.plot(X,lr.predict(X),color='red')
```

```python
#Lets apply GRADIENT DESCENT assuming slope is constant as m=44.27
#and lets assume that the starting value of b=0
y_pred=44.27*X+0
```

```python
plt.scatter(X,Y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
#Iteration 1
m=44.27
b=0

loss_slope=-2 * np.sum(Y-m*X.ravel()-b)
loss_slope
```

```python
# Lets take the learning rate as 0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=44.27*X+b
plt.scatter(X,Y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.plot(X,y_pred1,color='blue',label='b=7.11')

plt.legend()
plt.show()
```

```python
#Iteration 2
m=44.27
b=7.11

loss_slope=-2 * np.sum(Y-m*X.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=44.27*X+b
plt.scatter(X,Y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='yellow',label='b=8.53')
plt.plot(X,y_pred1,color='blue',label='b=7.11')
plt.plot(X,y_pred,color='green',label='b=0')

plt.legend()
plt.show()
```

```python
b=-100
m=44.27
lr=0.01

epoch=50

for i in range(epoch):
    loss_slope=-2*np.sum(Y-m*X.ravel()-b)
    b=b-(lr*loss_slope)

    y_pred=m*X+b

    plt.plot(X,y_pred)

plt.scatter(X,Y)
```
